# Naive bayers Algorithm

```python
#1. read the data provided in the same directory with name 'movies_sentiment_data.csv' and store it in df variable

import pandas as pd
df= pd.read_csv('IMDB Dataset.csv')


#2. print the shape of the data

df.shape

#3. print top 5 datapoints

df.head()
```

```python
#creating a new column "Category" which represent 1 if the sentiment is positive or 0 if it is negative
df['Category'] = df['sentiment'].apply(lambda x: 1 if x=='positive' else 0)
```

```python
df
```

```python
#check the distribution of 'Category' and see whether the Target labels are balanced or not.

df.Category.value_counts()
```

```python
#Do the 'train-test' splitting with test size of 20%
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.review, df.Category, test_size= 0.2)
```

```python
# using sklearn pipeline module create a classification pipeline to classify the movie review's positive or negative.
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline

clf= Pipeline([('Vectorizer', CountVectorizer()), ('nb', MultinomialNB() )])

# fit with X_train and y_train
clf.fit(X_train, y_train)
```

```python
# get the predictions for X_test and store it in y_pred
y_pred= clf.predict(X_test)
```

```python
# print the classfication report
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
# Class 0 (Negative Reviews):
# Precision (0.82): Out of all predicted negative reviews, 82% were actually negative.
# Recall (0.88): Out of all actual negative reviews, 88% were correctly identified as negative.
# F1-Score (0.85): Balances precision and recall, showing overall strong performance for this class.
# Support (4919): The model evaluated 4919 negative reviews in the dataset.


# Class 1 (Positive Reviews):
# Precision (0.87): Out of all predicted positive reviews, 87% were actually positive.
# Recall (0.81): Out of all actual positive reviews, 81% were correctly identified as positive.
# F1-Score (0.84): Indicates overall good performance for positive reviews.
# Support (5081): The model evaluated 5081 positive reviews.
```

```python
# Insights:
# The model performs equally well for both classes, with minor differences in precision, recall, and F1-scores.
# Precision for Class 1 (Positive) is higher, meaning fewer false positives for positive reviews.
# Recall for Class 0 (Negative) is higher, meaning fewer false negatives for negative reviews.
# The balanced dataset (equal support for both classes) likely contributed to the consistent metrics across both classes.
```

# Random Forest Classifier

```python
# using sklearn pipeline module create a classification pipeline to classify the movie review's positive or negative.
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
clf = Pipeline([
    ('Vectorizer', CountVectorizer(max_features=5000)),
    ('rf', RandomForestClassifier(n_estimators=50, n_jobs=-1))  # Using Random Forest Classifier
])

# fit with X_train and y_train
clf.fit(X_train, y_train)
```

```python
# get the predictions for X_test and store it in y_pred
y_pred= clf.predict(X_test)
```

```python
# print the classfication report
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

# KNeighbors Classifier

```python
# using sklearn pipeline module create a classification pipeline to classify the movie review's positive or negative.
from sklearn.feature_extraction.text import CountVectorizer
from  sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
```

```python
clf = Pipeline([
    ('Vectorizer', CountVectorizer()),
    ('rf', KNeighborsClassifier())  
])

# fit with X_train and y_train
clf.fit(X_train, y_train)
```

```python
# get the predictions for X_test and store it in y_pred
y_pred= clf.predict(X_test)
```

```python
# print the classfication report
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
